## Ejercicio 9
1. Escribe un programa que dada una variable titulo busque los libros que hay en la librería con ese título. Si no se encuentra debe devolver el mensaje `Ese no lo tengo, ¿mola?`
2. Pruena que el programa con uno de tus libros y con otro que no contenga tu librería.

`````python
def buscar_libro(libreria, titulo):
    resultados = []

    for libro in libreria:
        if libro["titulo"].lower() == titulo.lower():
            resultados.append(libro)

    if resultados:
        return resultados
    else:
        return "Ese no lo tengo, ¿mola?"

print(buscar_libro(libreria, "1984"))
print(buscar_libro(libreria, "Harry Potter"))
`````

## Ejercicio 10
Convierte el programa del ejercicio 2 en una función al que se le pase el valor a buscar como un argumento posicional

`````python
def encontrar_indices(lista, valor):
    indices = []

    for i, elemento in enumerate(lista):
        if elemento == valor:
            indices.append(i)

    return indices

ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un",
          "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]

print(encontrar_indices(ejer_3, "es"))
`````
