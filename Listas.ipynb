```python
asignaturas = ["Fisica", "Quimica","Matematicas", "Literatura"]

# Añadir elemento
asignaturas.append("Filosofia")
print(asignaturas)

# Añadir elementos en una posicion determinada
asignaturas.insert(1,"Educacion Fisica")
print(asignaturas)

# Quitar elementos
#asignaturas.clear()
#print(asignaturas)

# Quitar elemento a partir de su posicon
asignaturas.pop(1)
print(asignaturas)

#Quitar un elemento por su valor
asignaturas.remove("Filosofia")
print(asignaturas)

#asignaturas.remove("Cine")
#asignaturas.pop(123)

#Ordenar lista
asignaturas.sort()
print(asignaturas)
```
