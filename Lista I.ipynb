```python
degrees = [22, 34, 15, 26, 18, 22]

#Ordenar
degrees.sort()
print(degrees)

degrees.sort(reverse = True)
print(degrees)

#Contar el numero de elementos
print(len(degrees))

#INvertir la lista
degrees.reverse()
print(degrees)

#Ocurrencias de un elemento
print(degrees.count(22))
```
