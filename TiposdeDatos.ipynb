```python
print("Sumas y restas")
print(1 + 2)
print(1.0 + 2) # Fijate que cuando sumo flotar con int devuelve un float
print(1.0 - 2) # Siempre que opero float e int me dará float
```

```python
print("Mult y Divisiones")
print(2 * 2)
print(2.0 * 2)
print(2/2) # El tipo del resultado de una división siempre es float
      
```

```python
n = 56
print("¿Es", n, "par?")
resto = n % 2
print("El resto es ", resto)
if resto == 0:
    print("Es par")
else:
    print("Es impar")
```
